```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import math
```

### ReLU

```python
def max_relu(x):
  return max(x,0.0)

for i in [-10., -5., 0., 5., 10.]:
  print(max_relu(i)) #zeruje wartosci ujemne
```

```python
data = np.random.randn(50)
data = sorted(data)
data
```

```python
max_relu_data = np.array([max_relu(x) for x in data])
max_relu_data
```

```python
df = pd.DataFrame({'data':data, 'max_relu':max_relu_data})
df.head()
```

```python
px.line(df, x='data', y='max_relu', width=700, height=400, title='relu function')
```

### Sigmoid

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))

for i in [-5., -3., -1., 0., 1., 3., 5]:
  print(sigmoid(i))
```

```python
data = 3 * np.random.randn(50)
data = sorted(data)
data
```

```python
sigmoid_data = [sigmoid(x) for x in data]
sigmoid_data
```

```python
df = pd.DataFrame({'data':data, 'sigmoid':sigmoid_data})
df.head()
```

```python
px.line(df, x='data', y='sigmoid', width=700, height=400, title='sigmoid function')
```

### Tahn

```python
def tahn(x):
  return ((np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x)))

data = 2 * np.random.randn(100)
data = sorted(data)
data
```

```python
tahn_data = [tahn(x) for x in data]
tahn_data
```

```python
df = pd.DataFrame({'data':data, 'tahn':tahn_data})
df.head()
```

```python
px.line(df, x='data', y='tahn', width=700, height=400, title='tahn function')
```

### Softmax

```python
def softmax(x):
  e_x = np.exp(x)
  denominator = np.sum(e_x, axis=1)
  denominator = denominator[:, np.newaxis]
  return e_x / denominator

data = np.random.randn(4,5)
data
```

```python
result = softmax(data)
result
```
